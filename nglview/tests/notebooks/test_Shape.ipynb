```python
import nglview as nv

view = nv.NGLWidget()
view
```

```python
# adapted from http://arose.github.io/ngl/api/current/Shape.html

view.shape.add_sphere( [ 0, 0, 9 ], [ 1, 0, 0 ], 1.5 );
view.shape.add_ellipsoid( [ 6, 0, 0 ], [ 1, 0, 0 ], 1.5, [ 3, 0, 0 ], [ 0, 2, 0 ] );
view.shape.add_cylinder( [ 0, 2, 7 ], [ 0, 0, 9 ], [ 1, 1, 0 ], 0.5 );
view.shape.add_cone( [ 0, 2, 7 ], [ 0, 3, 3 ], [ 1, 1, 0 ], 1.5 );
view.shape.add_arrow( [ 1, 2, 7 ], [ 30, 3, 3 ], [ 1, 0, 1 ], 1.0 );
```

```python
view.render_image()
```

```python
view._display_image()
```
