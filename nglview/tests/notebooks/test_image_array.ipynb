```python
import nglview as nv
import pytraj as pt

traj = pt.datafiles.load_tz2()
view = nv.show_pytraj(traj)
view.clear()
view.add_licorice('backbone')
view
```

```python
view._hold_image = True
```

```python
import time

wait = 2.

view._image_array = []

for frame in range(0, 40, 10):
    view.frame = frame
    time.sleep(wait)
    view.render_image()
    time.sleep(wait)
```

```python
from IPython.display import Image, display

for im in view._image_array:
    display(Image(im))
```
