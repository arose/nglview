```python
import nglview as nv
from nglview.utils import get_colors_from_b64
import pytraj as pt

view = nv.demo()
view.clear()
view.add_cartoon(color='grey')
view._set_sync_camera()
view
```

```python
view.add_component('../../../../ngl/data/1crn.pdb')
```

```python
view.orientation = [[27.658191023394465, 32.24919855594635, 37.849483251571655],
 [70.90662963144484, -46.24593547229935, 49.10597974833395],
 [0.5589227770158776, 0.19227726217714627, -0.8066193549515858]]
```

```python
view.render_image()
```

```python
color_0 = get_colors_from_b64(view._image_data)
```

```python
view._display_image()
```

```python
view.superpose(components=[1,], ref=0)
```

```python
view.render_image()
```

```python
color_1 = get_colors_from_b64(view._image_data)
```

```python
view._display_image()
```

```python
assert color_0 != color_1
```

```python
view.superpose(components=[1,], ref=0, align=False)
```

```python
view.render_image()
```

```python
view._display_image()
```
